```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
```

```python
cosy = np.cos(x)
```

```python
siny = y.copy()
```

```python
plt.plot(x,siny,label = 'sin')
plt.plot(x,cosy,color = 'red',linestyle ='--',label = 'cos')
#plt.xlim(-5,15)
#plt.ylim(-2,2)
plt.axis([-1,11,-2,2])
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('xdk')
plt.show()
```

### Scatter Plot

```python
plt.scatter(x,siny)
plt.show()
```
